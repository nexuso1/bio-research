```python
import pandas as pd
import glob
```

```python
dfs = []

for path in glob.glob('../data/uniptm/*.csv'):
    dfs.append(pd.read_csv(path))
```

```python
dfs[2]
```

```python
residues = ['S', 'T', 'Y']
```

```python
import json
import numpy as np
from sklearn.model_selection import StratifiedKFold
```

```python
for df in dfs:
    _, length_bins = np.histogram(df['Length'], 10)
    length_labels = np.digitize(df['Length'], length_bins)
    df['length_class'] = length_labels

dfs[0]
```

```python
import ast

def site_fn(sites):
    x = ast.literal_eval(sites)
    if not isinstance(x, list):
        if isinstance(x, tuple):
            x = [str(site) for site in x] # str needed because of processing during training 
        else:
            x = [str(x)]

    return x

for df, residue in zip(dfs, residues):
    splits = []
    filtered = df[['Entry', 'Sequence', 'Position']].rename(columns={'Entry' : 'id', 'Sequence' : 'sequence', 'Position' : 'sites'})
    filtered['sites'] = filtered['sites'].apply(site_fn)
    filtered.to_json(f'../data/uniptm_info_{residue}.json')
    kfold = StratifiedKFold(5, shuffle=True, random_state=42)
    for train, test in kfold.split(df.index, df['length_class']):
        splits.append({'train' : df.index[train].to_list(), 'test' : df.index[test].to_list()})

    with open(f'../data/uniptm_splits_{residue}.json', 'w') as f:
        json.dump(splits, f)
```
